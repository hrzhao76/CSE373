```python
import os
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,10)
plt.rcParams["font.size"]=20
```

```python
path = "../Sea/SolomonResult/"
dir_list = os.listdir(path)
```

```python
dir_list
```

```python
fig1, ax = plt.subplots()
plt.xlabel("N")
plt.ylabel("Average runtime for addAll(s)")    
plt.title("addAll()")

fig2, ax2 = plt.subplots()
plt.xlabel("N")
plt.ylabel("Average runtime for allMatches(s)")
plt.title("allMatches()")
for csv in dir_list:
    Autocomplete = pd.read_csv(path+csv, header=None) 
    # plot
    
    ax.plot(Autocomplete[0], Autocomplete[1], linewidth=2.0, label =csv)
    ax2.plot(Autocomplete[0], Autocomplete[2], linewidth=2.0, label =csv)
    
ax.legend()
ax.set_yscale('log')
ax.set_ylim(10e-6,1)


ax2.legend()
ax2.set_yscale('log')
ax2.set_ylim(10e-7,10e-1)


# fig1.plt.legend()
# plt1
# 
# plt.xlabel("N")
    
```
